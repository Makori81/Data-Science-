```python
import matplotlib.pyplot as plt

labels = ["Java","Ruby","C++","Python"]
sections = [26.2,30.6,16.2,26.9]
colors =['c','red','g','y']
plt.pie(sections, labels=labels, colors=colors, startangle=45, autopct='%1.1f%%')

plt.title("PROGRAMMING LANGUAGES USAGE")
plt.show()
```
