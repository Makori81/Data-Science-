```python
import matplotlib.pyplot as plt
import numpy as np

Course = ["BSE","BBIT","BSCIT","BCS","BCM"]
students = [200,400,250,180,180]
colors = ['blue','y','green','red','cyan']

plt.figure(figsize=(8,5))

student_bars = plt.bar(Course, students, color=colors, width=0.8)

plt.plot()
plt.xlabel("COURSE")
plt.ylabel("NUMBER OF STUDENTS")
plt.title("INTAKE AS PER THE COURSES")
plt.yticks(np.arange(0,401,50))
plt.legend(student_bars, Course)
plt.tight_layout()
plt.show()
```
