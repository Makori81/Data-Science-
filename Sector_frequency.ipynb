```python
import matplotlib.pyplot as plt
import numpy as np

Sector = ["Utilities","Health care","Energy","Consumer staples","Communication services"]
Frequency = [50,470,280,130,130]
colors = ['purple','red','green','orange','violet']

plt.figure(figsize=(10,6))

plt.barh(Sector, Frequency, color=colors)

plt.plot()
plt.xlabel("frequency")
plt.ylabel("Sector")
plt.xticks(np.arange(0,500,100))
plt.tight_layout()
plt.show()
```
