```python
import matplotlib.pyplot as plt
import numpy as np

months=["07/2019","08/2019","09/2019","10/2019","11/2019"]
search_visitors=[50,53,59,56,62]
direct_visitors=[39,47,42,51,51]
sm_visitors=[70,80,90,87,92]

x_pos=np.arange(len(months))
width=0.25

plt.figure(figsize=(8,4))
ax=plt.gca()

search_bars=plt.bar(x_pos - width, search_visitors, width=width, label="Searches", color='blue')
direct_bars=plt.bar(x_pos, direct_visitors, width=width, label="Direct", color='pink')
sm_bars=plt.bar(x_pos + width, sm_visitors, width=width, label="Social Media", color='orange')

plt.bar_label(search_bars,padding=3)
plt.bar_label(direct_bars,padding=3)
plt.bar_label(sm_bars,padding=3)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.plot(1, 0, '>k', transform=ax.transAxes, clip_on=False, markersize=4)
ax.plot(0, 1, '^k', transform=ax.transAxes, clip_on=False, markersize=4)

ax.set_ylim(0,110)

plt.plot()
plt.xlabel("months",loc='right')
plt.ylabel("visitors\nin thousands",loc='top')
plt.title("Visitors by web traffic sources")
plt.xticks(x_pos,months)
plt.yticks(np.arange(0,101,20))
plt.legend(loc='upper center', bbox_to_anchor=(0.4,-0.15), ncol=3, frameon=False, handleheight=1.5)
plt.tight_layout()
plt.show()

```
