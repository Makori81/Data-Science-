```python
import matplotlib.pyplot as plt
import numpy as np

Age_group = ["0-10","11-20","21-30","31-40","41-50","51-60","61-70","71-80","80+"]
Male = [2,3,5,6,9,7,3,2,1]
Female = [1,1,2,2,3,1,1,0,1]

x_pos = np.arange(len(Age_group))
width=0.4
plt.figure(figsize=(12,6))

Male_bars = plt.bar(x_pos - width/2, Male, width=width, label="Male", color='orange')
Female_bars = plt.bar(x_pos + width/2, Female, width=width, label="Female", color='blue')

plt.plot()
plt.xlabel("Age Group")
plt.ylabel("Number of deaths")
plt.title("COVID-19 deaths by age and gender during lockdown")
plt.xticks(x_pos,Age_group)
plt.yticks(np.arange(0,10,2))
plt.legend()
plt.tight_layout()
plt.show()
```
