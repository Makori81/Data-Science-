```python
import matplotlib.pyplot as plt
import numpy as np

x = ["HP","DELL","LENOVO","ASUS","ACER"]
y = [22,33,13,15,17]

plt.figure(figsize=(7,6))

plt.bar(x, y, color='green', width=0.6)

plt.plot()
plt.xlabel("MODEL")
plt.ylabel("MARKET SHARE")
plt.title("DISTRIBUTION OF LAPTOP MODELS")
plt.xticks(x,rotation=45)
plt.yticks(np.arange(0,35,5))
plt.legend(['MY WORK'], frameon=True)
plt.grid(True)
plt.tight_layout()
plt.show()
```
